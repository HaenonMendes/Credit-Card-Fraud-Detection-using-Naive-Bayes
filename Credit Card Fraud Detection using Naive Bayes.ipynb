```python
# kernel density estimation (KDE) is a non-parametric way to estimate the probability density function of a random variable.
i = 1

fig, ax = plt.subplots(8,4,figsize=(18,30))
for column_header in column_head:    
    plt.subplot(8,4,i)
    sns.kdeplot(credit_fraud[column_header], bw = 0.4, label = "CreditFraud", shade=True, color="r", linestyle="--")
    sns.kdeplot(noncredit_fraud[column_header], bw = 0.4, label = "Non-Credit Fraud", shade=True, color= "y", linestyle=":")
    plt.title(column_header, fontsize=12)
    i = i + 1
plt.show();
```

```python
# Now we create the train and test datasets
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
creditcard_fraud['Amount_Norm'] = sc.fit_transform(creditcard_fraud['Amount'].values.reshape(-1,1))
```

```python
creditcard_fraud.head()
```

```python
creditcard_fraud = creditcard_fraud.drop(['Amount'], axis = 1)
```

```python
# Dropping the target label coloumns
X = creditcard_fraud.drop(['Class'],axis=1)
y = creditcard_fraud['Class']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Now we train the model
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.naive_bayes import GaussianNB 
Naive_Bayes_classifier = GaussianNB()
Naive_Bayes_classifier.fit(X_train, y_train)
```

```python
# Now we evaluate the model
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_predict_train = Naive_Bayes_classifier.predict(X_train)
y_predict_train
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot=True)
```

```python
# Predicting the Test set results
y_predict_test = Naive_Bayes_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict_test))
```

```python
# We can still improve the model
```

```python
X = creditcard_fraud.drop(['Time','V8','V13','V15','V20','V22','V23','V24','V25','V26','V27','V28','Class'], axis = 1)
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
Naive_Bayes_classifier = GaussianNB()
Naive_Bayes_classifier.fit(X_train, y_train)
y_predict = Naive_Bayes_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```
